## Problem 2

To get $\hat{p}$, we want $EL[Y=1|x] = EL[Y=0|x]$, then $L_{FP} * P(Y=0|x) + L_{TP}*P(Y=1|x) = L_{TN}*P(Y=0|x) + L_{FN}*P(Y=1|x)$, since $P(Y=1|x) = p, L_{TP} = L_{TN} = 0$, the equation becomes $L_{FP}*(1-p) = L_{FN}*p$ yielding $\hat{p} = \frac{L_{FP}}{L_{FN} + L_{FP}}$.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
fra_train = pd.read_csv("framingham_train.csv")
fra_test = pd.read_csv("framingham_test.csv")
```

```python
logreg = smf.logit(formula = 'TenYearCHD ~ male + age + education + currentSmoker + cigsPerDay+BPMeds+prevalentStroke+prevalentHyp+diabetes+totChol+sysBP+diaBP+BMI+heartRate+glucose',
                   data = fra_train).fit()
print(logreg.summary())
```

```python
fra_train2 = fra_train.copy() 
fra_train2['HighSchoolGED'] = (fra_train2['education'] == 'High school/GED').astype('int64')
fra_train2 = fra_train2.drop(columns='education')
fra_test2 = fra_test.copy() 
fra_test2['HighSchoolGED'] = (fra_test2['education'] == 'High school/GED').astype('int64')
fra_test2 = fra_test2.drop(columns='education')
```

```python
logreg1 = smf.logit(formula = 'TenYearCHD ~ male + age + HighSchoolGED + currentSmoker + cigsPerDay+BPMeds+prevalentStroke+prevalentHyp+diabetes+totChol+sysBP+diaBP+BMI+heartRate+glucose',
                   data = fra_train2).fit()

print(logreg1.summary())
```

```python
from sklearn.metrics import confusion_matrix

y_test = fra_test2['TenYearCHD']
y_prob = logreg1.predict(fra_test2)
y_pred = pd.Series([1 if x > 0.126 else 0 for x in y_prob], index=y_prob.index)

cm = confusion_matrix(y_test, y_pred)
print ("Confusion Matrix : \n", cm) 
```

```python
#Accuracy of the original model
(cm.ravel()[0]+cm.ravel()[3])/sum(cm.ravel())
```
