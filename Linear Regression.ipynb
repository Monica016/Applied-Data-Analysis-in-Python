```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

rogue = pd.read_csv('Rogue-242-Fall2021.csv')
rogue_train = rogue[rogue['Year'] <= 2015]
rogue_testa = rogue[(rogue['Year'] <= 2019) & (rogue['Year'] >=2016)]
rogue_testb = rogue[(rogue['Year'] <= 2021) & (rogue['Year'] >=2020)]
rogue.head(5)
```

```python
cols = ['Unemployment', 'RogueQueries', 'CPIEnergy', 'CPIAll']
X_train = rogue_train[cols]
Y_train = rogue_train['RogueSales']

import statsmodels.api as sm
X_train = sm.add_constant(X_train)

model = sm.OLS(Y_train, X_train).fit()
print(model.summary())
```

```python
from matplotlib import pyplot as plt
fig = sm.qqplot(model.resid, fit=True, line="45")
plt.show()

from statsmodels.graphics.plot_grids import scatter_ellipse
np.seterr(invalid='ignore')
fig = plt.figure()
scatter_ellipse(X_train, varnames=X_train.columns, fig=fig)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(df, columns):
    
    values = sm.add_constant(df[columns]).values  
    num_columns = len(columns)+1
    vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
    
    return pd.Series(vif[1:], index=columns)
```

```python
VIF(X_train, cols)
```

```python
cols2 = ['CPIAll', 'CPIEnergy', 'RogueQueries']
X_train2 = rogue_train[cols2]

import statsmodels.api as sm
X_train2 = sm.add_constant(X_train2)

model2 = sm.OLS(Y_train, X_train2).fit()
print(model2.summary())
```

```python
fig2 = sm.qqplot(model2.resid, fit=True, line="45")
plt.show()


fig2 = plt.figure()
scatter_ellipse(X_train2, varnames=X_train2.columns, fig=fig2)
plt.show()
```

```python
VIF(X_train2, cols2)
```

```python
def OSR2(model, df_train, df_test, dependent_var):   
    
    y_test = df_test[dependent_var]
    y_pred = model.predict(df_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(df_train[dependent_var]))**2)    
    
    return 1 - SSE/SST
```

```python
coltest = ['CPIAll', 'CPIEnergy', 'RogueQueries','RogueSales']
OSR2(model2, rogue_train[coltest], rogue_testa[coltest], 'RogueSales')
OSR2(model2, rogue_train[coltest], rogue_testb[coltest], 'RogueSales')
```
