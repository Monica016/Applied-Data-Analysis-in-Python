```python
import pandas as pd
raw = pd.read_csv('Bitcoin_tweets.csv')
```

```python
raw
```

```python
tweets = raw[['date','text','']]
```

```python
tweets['date'] = pd.Series([str(item).split()[0] for item in tweets['date']])
```

```python
tweets.head()
```

```python
from string import punctuation
def remove_punctuation_digit(document):
    document = document.lower()
    no_punct = ''.join([character for character in document if character not in punctuation])
    no_digit = ''.join([character for character in no_punct if not character.isdigit()])
    return no_digit
```

```python
tweets['text'] = pd.Series([str(item) for item in tweets['text']])
```

```python
tweets['text'] = tweets['text'].apply(remove_punctuation_digit)
```

```python
tweets
```

```python
def remove_nonenglish(document):
    encoded_string = document.encode("ascii", "ignore")
    decode_string = encoded_string.decode()
    return decode_string
```

```python
tweets['text'] = tweets['text'].apply(remove_nonenglish)
```

```python
tweets
```

```python
tweets = tweets.drop_duplicates(['text'])
```

```python
tweets
```

```python
def remove_n(document):
    re_n = document.replace("\n", "") 
    return re_n 
tweets['text'] = tweets['text'].apply(remove_n)
```

```python
tweets
```

```python
prices = pd.read_csv("BTC-USD.csv")

prices["Delta"] = prices["Close"] - prices["Open"]
dates = prices["Date"]
delta = prices["Delta"]
opens = prices["Open"]

sentiment = {}
for i in range(len(dates)):
    if abs(delta[i]) / opens[i] <= 0.01:
        sentiment[dates[i]] = "Neutral"
    else:
        if delta[i] > 0:
            sentiment[dates[i]] = "Positive"
        else:
            sentiment[dates[i]] = "Negative"
```

```python
sentiment
```

```python
tweets = tweets.drop([1347699,64943,137068,1885528])
```

```python
date_list = list(tweets["date"])
tweets["Tag"] = [sentiment[date] for date in date_list]
```

```python
tweets
```

```python
from sklearn.model_selection import train_test_split

tweets_train, tweets_test = train_test_split(tweets, test_size=0.3, random_state=88)
tweets_train.shape, tweets_test.shape
```

```python
from nltk.tokenize import word_tokenize
import nltk

tweets_train['text'] = tweets_train['text'].apply(word_tokenize)
tweets_test['text'] = tweets_test['text'].apply(word_tokenize)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(document):
    words = [word for word in document if not word in stop_words]
    return words
```

```python
tweets_train['text'] = tweets_train['text'].apply(remove_stopwords)
tweets_test['text'] = tweets_test['text'].apply(remove_stopwords)
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()

def stemmer(document):
    stemmed_document = [porter.stem(word) for word in document]
    return stemmed_document
```

```python
tweets_train['text'] = tweets_train['text'].apply(stemmer)
tweets_test['text'] = tweets_test['text'].apply(stemmer)
```

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer
from sklearn.feature_extraction.text import CountVectorizer

text_train = tweets_train['text'].apply(TreebankWordDetokenizer().detokenize)
countvec_train = CountVectorizer(min_df = 0.005)
sparse_dtm_train = countvec_train.fit_transform(text_train)
```

```python
text_test = tweets_test['text'].apply(TreebankWordDetokenizer().detokenize)
sparse_dtm_test = countvec_train.transform(text_test)
```

```python
dtm_train = pd.DataFrame(sparse_dtm_train.toarray(), columns=countvec_train.get_feature_names(),index=tweets_train.index)
dtm_test = pd.DataFrame(sparse_dtm_test.toarray(), columns=countvec_train.get_feature_names(),index=tweets_test.index)
```
